### **DIAGRAM TABEL**

```python
#membuat data sampel
data=[25,45,55,125,225]
kategori=['A','B','C','D','E']
```

```python
plt.bar(kategori,data)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')
```

```python
plt.bar(kategori,data,
        color='red',
        alpha=0.25)
#alpha untuk transparansi

#membuat grid
plt.grid(linestyle='-',
         linewidth=2,
         axis='y', #pilihan:x,y,both
         alpha=0.5)

for i in range(len(data)):
  plt.text(i, data[i], data[i], ha='center', va='bottom')

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot')
```

```python
plt.barh(kategori,data,
        color='red',
        alpha=0.25)
#alpha untuk transparansi

#membuat grid
plt.grid(linestyle='-',
         linewidth=2,
         axis='x', #pilihan:x,y,both
         alpha=0.5)

for i in range(len(data)):
  plt.text(i, data[i], data[i], ha='center', va='bottom')

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Horizontal Bar Plot')
```

```python
#membuat data sampel
data1=[25,85,74,40,60]
data2=[40,35,20,55,10]
kategori=['A','B','C','D','E']
```

```python
plt.bar(kategori,data1,label='Data 1')
plt.bar(kategori,data2,label='Data 2', bottom=data1)
plt.legend()

#membuat grid
plt.grid(linestyle='-',
         linewidth=2,
         axis='y', #pilihan;x,y,both
         alpha=0.5)

plt.xlabel('Jumlah')
plt.ylabel('Kategori')
plt.title('Contoh Stacked Bar Plot')
```

## **PIE CHART**

```python
data=[25,45,55,125,225]
kategori=['A','B','C','D','E']
```

```python
plt.pie(data,
         labels=kategori,
         autopct='%1.1f',
         startangle=180)
plt.title('Contoh Pie Plot')
```

# **EXPLODE PIE CHART**

```python
data=[25,45,55,125,225]
kategori=['A','B','C','D','E']
warna=['pink','cyan','yellowgreen','skyblue','purple']
explode_var=[0.,0.,0.2,0.,0.,]
```

```python
plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f',
        startangle=90)

plt.legend()
plt.title('Contoh Explode Pie Plot')
```
