```python
import pandas as pd
import matplotlib.pyplot as plt

# membaca data
df = pd.read_excel("StudentPerformanceFactors12.csv.xlsx")

# melihat data awal
print(df.head())
```

```python
gender_counts = df['Gender'].value_counts()

plt.figure()
gender_counts.plot(kind='bar')

plt.title("Distribusi Gender Siswa")
plt.xlabel("Gender")
plt.ylabel("Jumlah Siswa")

plt.show()
```

```python
motivation_counts = df['Motivation_Level'].value_counts()

plt.figure()
plt.pie(motivation_counts, labels=motivation_counts.index, autopct='%1.1f%%')

plt.title("Distribusi Motivation Level Siswa")

plt.show()
```

```python
plt.figure()
plt.hist(df['Exam_Score'], bins=20)

plt.title("Distribusi Nilai Ujian")
plt.xlabel("Nilai Ujian")
plt.ylabel("Jumlah Siswa")

plt.show()
```

```python
avg_score = df.groupby('Parental_Involvement')['Exam_Score'].mean()

plt.figure()
avg_score.plot(kind='bar')

plt.title("Rata-rata Nilai Berdasarkan Parental Involvement")
plt.xlabel("Parental Involvement")
plt.ylabel("Rata-rata Exam Score")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel("StudentPerformanceFactors12.csv.xlsx")

fig, ax = plt.subplots(2, 2, figsize=(12,8))

# grafik 1
gender = df['Gender'].value_counts()
ax[0,0].bar(gender.index, gender.values)
ax[0,0].set_title("Distribusi Gender")

# grafik 2
motivation = df['Motivation_Level'].value_counts()
ax[0,1].pie(motivation, labels=motivation.index, autopct='%1.1f%%')
ax[0,1].set_title("Motivation Level")

# grafik 3
ax[1,0].hist(df['Exam_Score'], bins=20)
ax[1,0].set_title("Distribusi Nilai")

# grafik 4
parent = df.groupby('Parental_Involvement')['Exam_Score'].mean()
ax[1,1].bar(parent.index, parent.values)
ax[1,1].set_title("Rata-rata Nilai vs Parental Involvement")

plt.tight_layout()
plt.show()
```
