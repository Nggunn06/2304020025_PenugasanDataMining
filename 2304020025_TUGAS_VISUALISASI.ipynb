```python
#membuat multiple subplot
plt.figure(figsize=(20, 45))


#Jam Belajar vs Nilai Ujian
plt.subplot(10, 2, 1)
plt.bar(a, t, color='pink')
plt.grid()
plt.xlabel('Jam Belajar')
plt.ylabel('Nilai Ujian')
plt.title('Hubungan antara Jam Belajar dengan Nilai Ujian')

#Kehadiran vs Nilai Ujian
plt.subplot(10, 2, 2)
plt.bar(b, t, color='lightblue')
plt.grid()
plt.xlabel('Kehadiran')
plt.ylabel('Nilai Ujian')
plt.title('Hubungan antara Kehadiran dengan Nilai Ujian')

# Keterlibatan Orang Tua vs Nilai Ujian
plt.subplot(10, 2, 3)
# Hitung rata-rata untuk setiap kategori
c_means = t.groupby(c).mean()
plt.bar(c_means.index, c_means.values, color='lightgreen')
plt.grid()
plt.xlabel('Keterlibatan Orang Tua')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Keterlibatan Orang Tua dengan Nilai Ujian')

# Akses ke Resources vs Nilai Ujian
plt.subplot(10, 2, 4)
d_means = t.groupby(d).mean()
plt.bar(d_means.index, d_means.values, color='yellow')
plt.grid()
plt.xlabel('Akses ke Resources')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Akses Resources dengan Nilai Ujian')

#Ekstrakurikuler vs Nilai Ujian

plt.subplot(10, 2, 5)
e_means = t.groupby(e).mean()
plt.bar(e_means.index, e_means.values, color='orange')
plt.grid()
plt.xlabel('Ekstrakurikuler')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Ekstrakurikuler dengan Nilai Ujian')

# Jam Tidur vs Nilai Ujian
plt.subplot(10, 2, 6)
plt.bar(f, t, color='purple')
plt.grid()
plt.xlabel('Jam Tidur')
plt.ylabel('Nilai Ujian')
plt.title('Hubungan antara Jam Tidur dengan Nilai Ujian')

# Nilai Sebelumnya vs Nilai Ujian
plt.subplot(10, 2, 7)
plt.bar(g, t, color='red')
plt.grid()
plt.xlabel('Nilai Sebelumnya')
plt.ylabel('Nilai Ujian')
plt.title('Hubungan antara Nilai Sebelumnya dengan Nilai Ujian')

# Motivasi vs Nilai Ujian
plt.subplot(10, 2, 8)
h_means = t.groupby(h).mean()
h_means = h_means[['Low', 'Medium', 'High']]  # Urutkan
plt.bar(h_means.index, h_means.values, color=['red', 'yellow', 'green'])
plt.grid()
plt.xlabel('Motivasi')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Motivasi dengan Nilai Ujian')

# Akses Internet vs Nilai Ujian
plt.subplot(10, 2, 9)
i_means = t.groupby(i).mean()
plt.bar(i_means.index, i_means.values, color='cyan')
plt.grid()
plt.xlabel('Akses Internet')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Akses Internet dengan Nilai Ujian')

# BARIS 5: Sesi Bimbingan vs Nilai Ujian
plt.subplot(10, 2, 10)
plt.bar(j, t, color='magenta')
plt.grid()
plt.xlabel('Sesi Bimbingan')
plt.ylabel('Nilai Ujian')
plt.title('Hubungan antara Sesi Bimbingan dengan Nilai Ujian')

# BARIS 6: Pendapatan Keluarga vs Nilai Ujian
plt.subplot(10, 2, 11)
k_means = t.groupby(k).mean()
k_means = k_means[['Low', 'Medium', 'High']]
plt.bar(k_means.index, k_means.values, color=['red', 'yellow', 'green'])
plt.grid()
plt.xlabel('Pendapatan Keluarga')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Pendapatan Keluarga dengan Nilai Ujian')

# BARIS 6: Kualitas Guru vs Nilai Ujian
plt.subplot(10, 2, 12)
l_means = t.groupby(l).mean()
l_means = l_means[['Low', 'Medium', 'High']]
plt.bar(l_means.index, l_means.values, color=['red', 'yellow', 'green'])
plt.grid()
plt.xlabel('Kualitas Guru')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Kualitas Guru dengan Nilai Ujian')

# Tipe Sekolah vs Nilai Ujian
plt.subplot(10, 2, 13)
m_means = t.groupby(m).mean()
plt.bar(m_means.index, m_means.values, color=['blue', 'orange'])
plt.grid()
plt.xlabel('Tipe Sekolah')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Tipe Sekolah dengan Nilai Ujian')

# Pengaruh Teman vs Nilai Ujian
plt.subplot(10, 2, 14)
n_means = t.groupby(n).mean()
n_means = n_means[['Negative', 'Neutral', 'Positive']]
plt.bar(n_means.index, n_means.values, color=['red', 'yellow', 'green'])
plt.grid()
plt.xlabel('Pengaruh Teman')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Pengaruh Teman dengan Nilai Ujian')

# Aktivitas Fisik vs Nilai Ujian
plt.subplot(10, 2, 15)
plt.bar(o, t, color='brown')
plt.grid()
plt.xlabel('Aktivitas Fisik')
plt.ylabel('Nilai Ujian')
plt.title('Hubungan antara Aktivitas Fisik dengan Nilai Ujian')

# Kesulitan Belajar vs Nilai Ujian
plt.subplot(10, 2, 16)
p_means = t.groupby(p).mean()
plt.bar(p_means.index, p_means.values, color=['green', 'red'])
plt.grid()
plt.xlabel('Kesulitan Belajar')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Kesulitan Belajar dengan Nilai Ujian')

# Pendidikan Orang Tua vs Nilai Ujian
plt.subplot(10, 2, 17)
q_means = t.groupby(q).mean()
plt.bar(range(len(q_means)), q_means.values, tick_label=q_means.index, color='teal')
plt.grid()
plt.xlabel('Pendidikan Orang Tua')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Pendidikan Orang Tua dengan Nilai Ujian')

# Jarak Rumah vs Nilai Ujian
plt.subplot(10, 2, 18)
r_means = t.groupby(r).mean()
r_means = r_means[['Near', 'Moderate', 'Far']]
plt.bar(r_means.index, r_means.values, color=['green', 'yellow', 'red'])
plt.grid()
plt.xlabel('Jarak Rumah')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Jarak Rumah dengan Nilai Ujian')

# Gender vs Nilai Ujian
plt.subplot(10, 2, 19)
s_means = t.groupby(s).mean()
plt.bar(s_means.index, s_means.values, color=['lightblue', 'pink'])
plt.grid()
plt.xlabel('Gender')
plt.ylabel('Rata-rata Nilai Ujian')
plt.title('Hubungan Gender dengan Nilai Ujian')

# BARIS 10: Distribusi Nilai Ujian
plt.subplot(10, 2, 20)
plt.hist(t, bins=30, color='skyblue', edgecolor='black')
plt.grid()
plt.xlabel('Nilai Ujian')
plt.ylabel('Frekuensi')
plt.title('Distribusi Nilai Ujian')

plt.tight_layout()
plt.show()
```

**PIE CHART**

```python
#membuat multiple subplot untuk pie chart
plt.figure(figsize=(20, 25))

# PIE 1: Parental Involvement
plt.subplot(4, 4, 1)
data = c.value_counts().values
kategori = c.value_counts().index
warna = ['#ff9999', '#66b3ff', '#99ff99']
explode_var = [0, 0, 0.1]  # High diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Parental Involvement')
plt.axis('equal')

# PIE 2: Access to Resources
plt.subplot(4, 4, 2)
data = d.value_counts().values
kategori = d.value_counts().index
warna = ['#ff9999', '#66b3ff', '#99ff99']
explode_var = [0.1, 0, 0]  # Low diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Access to Resources')
plt.axis('equal')

# PIE 3: Extracurricular Activities
plt.subplot(4, 4, 3)
data = e.value_counts().values
kategori = e.value_counts().index
warna = ['#ff9999', '#66b3ff']
explode_var = [0, 0.1]  # Yes diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Extracurricular Activities')
plt.axis('equal')

# PIE 4: Motivation Level
plt.subplot(4, 4, 4)
# Urutkan
h_counts = h.value_counts()
h_counts = h_counts[['Low', 'Medium', 'High']]
data = h_counts.values
kategori = h_counts.index
warna = ['#ff9999', '#66b3ff', '#99ff99']
explode_var = [0, 0, 0.1]  # High diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Motivation Level')
plt.axis('equal')

# PIE 5: Internet Access
plt.subplot(4, 4, 5)
data = i.value_counts().values
kategori = i.value_counts().index
warna = ['#99ff99', '#ff9999']
explode_var = [0, 0.1]  # No diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Internet Access')
plt.axis('equal')

# PIE 6: Family Income
plt.subplot(4, 4, 6)
k_counts = k.value_counts()
k_counts = k_counts[['Low', 'Medium', 'High']]
data = k_counts.values
kategori = k_counts.index
warna = ['#ff9999', '#66b3ff', '#99ff99']
explode_var = [0, 0, 0.1]  # High diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Family Income')
plt.axis('equal')

# PIE 7: Teacher Quality
plt.subplot(4, 4, 7)
l_counts = l.value_counts()
l_counts = l_counts[['Low', 'Medium', 'High']]
data = l_counts.values
kategori = l_counts.index
warna = ['#ff9999', '#66b3ff', '#99ff99']
explode_var = [0.1, 0, 0]  # Low diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Teacher Quality')
plt.axis('equal')

# PIE 8: School Type
plt.subplot(4, 4, 8)
data = m.value_counts().values
kategori = m.value_counts().index
warna = ['#66b3ff', '#99ff99']
explode_var = [0, 0.1]  # Private diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('School Type')
plt.axis('equal')

# PIE 9: Peer Influence
plt.subplot(4, 4, 9)
n_counts = n.value_counts()
n_counts = n_counts[['Negative', 'Neutral', 'Positive']]
data = n_counts.values
kategori = n_counts.index
warna = ['#ff9999', '#66b3ff', '#99ff99']
explode_var = [0.1, 0, 0]  # Negative diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Peer Influence')
plt.axis('equal')

# PIE 10: Learning Disabilities
plt.subplot(4, 4, 10)
data = p.value_counts().values
kategori = p.value_counts().index
warna = ['#99ff99', '#ff9999']
explode_var = [0, 0.1]  # Yes diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Learning Disabilities')
plt.axis('equal')

# PIE 11: Parental Education Level
plt.subplot(4, 4, 11)
q_counts = q.value_counts()
data = q_counts.values
kategori = q_counts.index
warna = ['#ff9999', '#66b3ff', '#99ff99']
explode_var = [0, 0, 0.1]  # Postgraduate diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Parental Education')
plt.axis('equal')

# PIE 12: Distance from Home
plt.subplot(4, 4, 12)
r_counts = r.value_counts()
r_counts = r_counts[['Near', 'Moderate', 'Far']]
data = r_counts.values
kategori = r_counts.index
warna = ['#99ff99', '#66b3ff', '#ff9999']
explode_var = [0, 0.1, 0]  # Moderate diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Distance from Home')
plt.axis('equal')

# PIE 13: Gender
plt.subplot(4, 4, 13)
data = s.value_counts().values
kategori = s.value_counts().index
warna = ['lightblue', 'pink']
explode_var = [0, 0.1]  # Female diexplode

plt.pie(data,
        labels=kategori,
        colors=warna,
        explode=explode_var,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Gender')
plt.axis('equal')
```
